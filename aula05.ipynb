```python
import pandas as pd

df = pd.read_csv('linguagens.csv')

#ano1 = df["Ano"] == 1 #pega cada linha onde ano = 1, aí mostra true onde é, false onde não é
ano1 = df[df["Ano"] == 1] #retorna apenas as linhas que são true
ano2 = df[df["Ano"] == 2] #retorna apenas as linhas que são true

valores_ano1 = ano1["Valor"].sum() #pega a coluna valor do dataframe ano1
valores_ano2 = ano2["Valor"].sum() #pega a coluna valor do dataframe ano2

print("Participacao das linguagens:")
print(f"Ano 1: {valores_ano1}") #soma os valores da coluna valor do dataframe ano1
print(f"Ano 2: {valores_ano2}") #soma os valores da coluna valor do dataframe ano2

print() #pula uma linha

print("Crescimento das linguagens:")
#agora precisamos fazer um for para calcular a taxa de crescimento de cada linguagem, porque antes o pandas
#já tinha uma função de soma, mas ele não tem uma função de taxa de crescimento
linguagens = df["Linguagem"].unique() #pega todas as linguagens(sem repetição) e transforma em uma lista
taxas_crescimento = {}
for linguagem in linguagens: #vai percorrer cada elemento da lista e comparar com a tabela na coluna de linguagens
    #além disso, também filtra todas as colunas que são de ano 1
    #o .sum(), nesse caso, ele não vai pegar nada, pois ele só está recebendo uma linha na filtragem
    #como no csv tem apenas uma linguagem por ano, só vai ter um valor por ano, logo a soma vai ser ele mesmo
    valor_ano1 = df[(df["Linguagem"] == linguagem) & (df["Ano"] == 1)]["Valor"].sum() #pega o valor da linguagem no ano 1
    valor_ano2 = df[(df["Linguagem"] == linguagem) & (df["Ano"] == 2)]["Valor"].sum() #pega o valor da linguagem no ano 2
     
    taxa = (valor_ano2 - valor_ano1) / valor_ano1 * 100 #calcula a taxa de crescimento
    taxas_crescimento[linguagem] = taxa #adiciona a taxa de crescimento na lista, onde a chave é a linguagem e o valor é a taxa de crescimento
    
    #linguagem recebe todos os elementos da lista linguagens, logo ele recebe o nome das linguagens
    print(f"{linguagem}: crescimento = {taxa:.2f}%") #mostra a taxa de crescimento com 2 casas decimais
    
print() #pula uma linha

print("Linguagens que tiveram crescimento positivo:")
for linguagem in linguagens: #vai percorrer cada elemento da lista e comparar com a tabela na coluna de linguagens
    #além disso, também filtra todas as colunas que são de ano 1
    #o .sum(), nesse caso, ele não vai pegar nada, pois ele só está recebendo uma linha na filtragem
    #como no csv tem apenas uma linguagem por ano, só vai ter um valor por ano, logo a soma vai ser ele mesmo
    valor_ano1 = df[(df["Linguagem"] == linguagem) & (df["Ano"] == 1)]["Valor"].sum() #pega o valor da linguagem no ano 1
    valor_ano2 = df[(df["Linguagem"] == linguagem) & (df["Ano"] == 2)]["Valor"].sum() #pega o valor da linguagem no ano 2
    
    taxa_crescimento = (valor_ano2 - valor_ano1) / valor_ano1 * 100 #calcula a taxa de crescimento
    
    if taxa_crescimento > 0: #se a taxa de crescimento for maior que 0, ou seja, positivo
        # MOSTRA O CABEÇALHO UMA VEZ ANTES DOS DADOS
        print("Linguagem  Ano  Valor")
        
        # Mostra linha do Ano 1
        print(f"{linguagem:>9}  {1:>3}  {valor_ano1:>5.2f}")
        
        # Mostra linha do Ano 2  
        print(f"{linguagem:>9}  {2:>3}  {valor_ano2:>5.2f}")
        
        print(f"Crescimento: {taxa_crescimento:.2f}%")
        
        
print() #pula uma linha

print("Projeções para daqui a dois anos:") #Valor Futuro = Valor Presente × (1 + taxa)ⁿ

projecoes = {}
for linguagem in linguagens: 
    valor_ano2 = df[(df["Linguagem"] == linguagem) & (df["Ano"] == 2)]["Valor"].sum() #valor presente
    taxa = taxas_crescimento[linguagem] / 100  # pega a taxa de crescimento de cada linguagem do dicionário e divide por 100 para transformar em decimal
    
    # Fórmula: Valor Futuro = Valor Presente × (1 + taxa)²
    projecao = valor_ano2 * ((1 + taxa) ** 2)
    projecoes[linguagem] = projecao #linguagem funciona como um filtro. Adiciona no dicionário a linguagem como chave e a projeção como valor
    
    print(f"{linguagem}: {projecao:.2f}")

# Encontra a linguagem com maior projeção
linguagem_mais_popular = max(projecoes, key=projecoes.get)
print(f"\nA linguagem mais popular será: {linguagem_mais_popular}")
        

        
        
```
