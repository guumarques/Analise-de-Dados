# Seaborn - Usado para montar gráficos

```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt

#dataset embutido no seaborn
tips = sns.load_dataset("tips")
print(tips.head())
print(tips.info())
```

# Histograma(distribuição de valores)

```python
sns.histplot(data=tips, x="total_bill", bins = 20)
plt.show()
```

# Gráfico de dispersão(relação entre variáveis)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()

print(tips['total_bill'].corr(tips['tip']))  # Correlação entre total_bill e tip(total da conta e gorjeta)
```

# Scatterplot com linha de regressão

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(# linear model plot
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "sex", #cores diferentes para homens e mulheres
    scatter_kws={"alpha": 0.6}, #deixa os pontos mais transparentes
    line_kws={"linewidth": 2} #deixa a linha mais grossa
)

plt.title("Relação entre total da conta e gorjeta, por sexo")
plt.show()
```

# Heatmap(mapa de calor)

```python
#é um tipo de gráfico que mostra valores usando cores em vez de números. Muito usado para visualizar correlações entre várias variáveis.

colunas_selecionadas = ["total_bill", "tip", "size"]
corr = tips[colunas_selecionadas].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

# Média

```python
import numpy as np

#é o valor equilibrado, como se todos tivessem contribuído igualmente
#é sensível a valores extremos(outliers)
data1 = np.array([2, 3, 5, 7, 8, 10, 11])
print(np.mean(data1))  # Média

data2 = np.array([2, 3, 5, 7, 8, 10, 11, 50])
print(np.mean(data2))  # Mediana
```

# Mediana

```python
import numpy as np

#se for impar, ele pega o valor central
#se for par, ele pega a média dos dois valores centrais

data1 = np.array([2, 3, 5, 7, 8, 10, 11])
print(np.median(data1))  # Mediana

data2 = np.array([2, 3, 5, 7, 8, 10, 11, 50])
print(np.median(data2))  # Mediana
```

# Quartis

```python
import numpy as np

data = np.array([2, 3, 5, 7, 8, 10, 11])

q1 = np.quantile(data, 0.25)  
q3 = np.quantile(data, 0.75) 

print("1º Quartil (25%):", q1)
print("3º Quartil (75%):", q3)
```

# Intervalo interquartil(IQR)

```python
#Mede o intervalos dos 50% centrais dos dados e indica o quanto esse porção central se dispersa

IQR = q3 - q1
print("Intervalo Interquartil (IQR):", IQR)
```
