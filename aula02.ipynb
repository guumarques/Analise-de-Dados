```python
#operadores

a = 2
b = 3

soma = a + b
sub = a - b
mult = a * b
div = a / b
div_inteira = a // b
potencia = a ** b
resto = a % b

print(f'soma = {soma}')
print(f'subtração = {sub}')
print(f'multiplicação = {mult}')
print(f'divisão = {div}')
print(f'divisão inteira = {div_inteira}')
print(f'potência = {potencia}')
print(f'resto = {resto}')
```

```python
#Bibliotecas

import math

x = float(input())

print(math.trunc(x)) #arredonda para baixo

print(math.ceil(x)) #arredonda para cima

print(math.floor(x)) #arredonda para baixo

print(round(x)) #arredonda para cima quando par e para baixo quando ímpar(lado esquerdo da vírgula)
```

```python
#cadeia de caracteres

var = 'Hello World'

print(var[0]) #mostra a letra da posição zero
print(var[6:11]) #mostra as letras da posição seis até a onze
print(var[6:]) #mostra da posição seis até o final
print(var[0:11:2]) #vai de zero até o final, porém de dois em dois
```

```python
#funções úteis

var = 'Hello World'

print(len(var)) #comprimento do var
print(var.count('o')) #procura padrões dentro de uma string(nesse caso, procura quantas letra O tem na palavra Hello World)
print(var.count('l', 0,5)) #conta quantas letras L tem na palavra Hello
print(var.find('lo')) #posição onde está "lo"
print('World' in var) #verifica se a palavra world está dentro de var
var.replace('World', 'Python') #substitui a palavra World por python
print(var.upper()) #deixa tudo em letra maiúscula
print(var.lower()) #deixa tudo em letra minúscula
```
