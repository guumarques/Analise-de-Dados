```python
import pandas as pd
df = pd.read_csv('dirtydata.csv')
print(df.to_string())
```

# Eliminando linhas com dados faltantes

```python
df1 = df.dropna()
print(df1.to_string())

#por padrão, o método dropna retorna um novo dataframe, e não muda o original. 
#se você quiser mudar o original, use o parâmetro inplace=True
```

# Substituindo os valores faltantes


```python
df2 = df.copy()
df2.fillna(130, inplace=True)
print(df2.to_string())

#não parece bom, pois vai substituir todos os valores faltantes por 130, o que não faz sentido para a coluna 'Nome', data
```

# Substituindo apenas colunas específicas

```python
df3 = df.copy()
df3.fillna({'Calories': 130}, inplace=True)
print(df3.to_string())
```

# Removendo linhas duplicadas

```python
df.drop_duplicates(inplace=True)
print(df.to_string())

#mexe diretamente no df original
```

# Linhas com o formato errado

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce') #se não conseguir converter, coloca NaT

df =  df.dropna(subset=['Date']) #remove linhas onde a coluna Date apresenta NaT
print(df.to_string())
```

# Exercícios

```python
import pandas as pd

arquivo = input()

df = pd.read_csv(arquivo)

# Identifica as linhas com "?"
linhas_erradas = df[(df['Sub_metering_1'] == '?') | 
                    (df['Sub_metering_2'] == '?') | 
                    (df['Sub_metering_3'] == '?')
                    ].index

# Remove essas linhas do DataFrame
df.drop(linhas_erradas, inplace=True)

# Converte para valores numéricos
df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']] = df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].apply(pd.to_numeric)

# Calcula médias
media_medidor1 = df['Sub_metering_1'].mean()
media_medidor2 = df['Sub_metering_2'].mean()
media_medidor3 = df['Sub_metering_3'].mean()

# Calcula máximos
max_medidor1 = df['Sub_metering_1'].max()
max_medidor2 = df['Sub_metering_2'].max()
max_medidor3 = df['Sub_metering_3'].max()

# Imprime no formato pedido
print("Media Sub1: ", media_medidor1, "\n")
print("Media Sub2: ", media_medidor2, "\n")
print("Media Sub3: ", media_medidor3, "\n")
print("Maximo Sub1: ", max_medidor1, "\n")
print("Maximo Sub2: ", max_medidor2, "\n")
print("Maximo Sub3: ", max_medidor3, "\n")
```
