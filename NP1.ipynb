```python
import pandas as pd

# Carregar o arquivo CSV
nome_arquivo = input()
df = pd.read_csv(nome_arquivo)

# Removendo linhas duplicadas e com dados faltantes
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)

# Lendo limiar
X = int(input())

# Quantidade de livros que custam mais de US$ X
precos = df['Price']
nLivrosCaros = precos[precos > X].count()
print(nLivrosCaros)
```

```python
import pandas as pd

# Carregar o arquivo CSV
nome_arquivo = input()
df = pd.read_csv(nome_arquivo)

# Removendo linhas duplicadas e com dados faltantes
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)

# Preço médio dos livros por ano
preco_medio_ano = df.groupby(df['Year'])['Price'].mean()
print(preco_medio_ano)
```

```python
import pandas as pd

# Carregar o arquivo CSV
nome_arquivo = input()
df = pd.read_csv(nome_arquivo)

# Removendo linhas duplicadas e com dados faltantes
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)

# Lendo X
X = float(input())

# Título e preço do livro mais caro cuja avaliação é inferior a X
low_rating = df[df["User Rating"] < X] # livros cuja avaliação é inferior a X
max_price = low_rating.loc[low_rating["Price"].idxmax()] # localizando livro mais caro
print(max_price[["Name","Price"]])
```

```python
import pandas as pd

# Carregar o arquivo CSV
nome_arquivo = input()
df = pd.read_csv(nome_arquivo)

# Removendo linhas duplicadas e com dados faltantes
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)

# Gênero com maior média de avaliações
media_avaliacao_genero = df.groupby('Genre')['User Rating'].mean()
genero_maior_media = media_avaliacao_genero.idxmax()
maior_media = media_avaliacao_genero.max()
print(genero_maior_media, maior_media)
```
