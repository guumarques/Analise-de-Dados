Faça um programa que leia o nome de um arquivo csv e o carrege para um dataframe. 

Este arquivo contém dados de compras, com as seguintes colunas: 

Age, Gender, Item Purchased, Category, Purchase Amount (USD), Location, Size, Color, Season, Review Rating, Subscription Status, Shipping Type, Discount Applied, Promo Code Used, Previous Purchases, Payment Method, Frequency of Purchases.

As colunas Age, Purchase Amount (USD), Previous Purchases são do tipo int.

As coluna Review Rating é do tipo float. 

As demais são do tipo string.

Neste dataset existem dados faltas e linhas duplicadas.

Faça um programa que remova as linhas com dados faltantes e as linhas duplicadas e retorne a média da coluna Age

```python
import pandas as pd

arquivo = "shopping_trends_dirty.csv"

df = pd.read_csv(arquivo, index_col=0)
df.dropna(inplace=True)   #elimina linhas com dados faltantes do dataframe original
df.drop_duplicates(inplace=True) #remove as duplicatas do dataframe original

media_age = df['Age'].mean() #média das colunas Age

print(media_age)

```

Faça um programa que remova as linhas com dados faltantes e as linhas duplicadas e retorne a quantidade de clientes que gastaram mais que US$80 em compras

```python
import pandas as pd

arquivo = "shopping_trends_dirty.csv" #importando o nome do arquivo
df = pd.read_csv(arquivo, index_col=0)

df.dropna(inplace=True) #remove linhas com dados faltantes 
df.drop_duplicates(inplace=True) #remove duplicatas do dataframe original

maior_oitenta = df[df['Purchase Amount (USD)'] > 80]['Purchase Amount (USD)'] #filtra a coluna de compras com valores maiores que 80

print(maior_oitenta.count()) #conta quantas linhas foram filtradas
```

Faça um programa que remova as linhas com dados faltantes e as linhas duplicadas e retorne a porcentagem de vendas do item menos vendido da loja

```python
import pandas as pd

arquivo = "shopping_trends_dirty.csv"

df = pd.read_csv(arquivo)

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

#agrupa os itens comprados com a quantidade comprada
quantidade = df['Item Purchased'].value_counts() #conta quantas vezes cada item aparece na tabela

quantidade_menos_vendido = quantidade.min() #pega o item que tem a menor venda
total_vendas = quantidade.sum() #soma todas as quantidade de vendas de todos os itens da loja
    
porcentagem = (quantidade_menos_vendido/total_vendas) * 100

print(porcentagem)


#valor esperado -> 3.184386235233693
```

Faça um programa que remova as linhas com dados faltantes e as linhas duplicadas e retorne a porcentagem de compras realizadas com o método de pagamento

```python
import pandas as pd

arquivo = "shopping_trends_dirty.csv"

df = pd.read_csv(arquivo)

df.dropna(inplace=True) #elimina todas as linhas com dados faltantes
df.drop_duplicates(inplace=True) #elimina as duplicatas

todas_compras = df['Payment Method'] #pega todos os elementos da coluna de pagamento
quantidade_todas_compras = len(todas_compras) #conta quantos elementos a coluna tem

compras_venmo = df.loc[df['Payment Method'] == 'Venmo']['Payment Method'] #filtra todos os elementos da coluna de pagamento que possuemo nome "Venmo"
quantidade_compras_venmo = len(compras_venmo) #conta quantos resultados foram obtidos na consulta anterior

porcentagem = (quantidade_compras_venmo/quantidade_todas_compras) * 100

print(porcentagem)
```

Faça um programa que remova as linhas com dados faltantes e as linhas duplicadas e retorne o valor total de compras(Purchase Amount) por método de pagamento(Payment Method)

```python
import pandas as pd

arquivo = "shopping_trends_dirty.csv"

df = pd.read_csv(arquivo)

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Agrupa pelo método de pagamento e soma o valor total das compras
total_por_metodo = df.groupby('Payment Method')['Purchase Amount (USD)'].sum() #monta uma série do pandas, filtrado pelo groupy

# Exibe o resultado
print(total_por_metodo)
```

# Questão 3 da prova 1

Retorne o nome e preço do livro mais caro que possuir a avaliação menor que x

```python
import pandas as pd

arquivo = input()

df = pd.read_csv(arquivo)

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

numero = float(input())

ratings = df[df['Ratings'] < numero]

livro_mais_caro = ratings.loc[ratings['Price'].idxmax(), ['Book Name', 'Price']]

print(livro_mais_caro)
```
