# Tupla

```python
nomes = ('Goku', 'Vegeta', 'Trunks', 'Gohan') #tupla(coleção de items ordenada e imutável)
print(nomes) #printa a tupla completa
print(nomes[:2]) #printa os dois primeiros itens da tupla
print(len(nomes)) #printa o tamanho da tupla
print(sorted(nomes)) #printa a tupla em ordem alfabética
```

# Lista

```python
nomes = ['Goku', 'Vegeta', 'Trunks', 'Gohan'] #lista(coleção de items ordenada e mutável)
#nomes.append('Bulma') #adiciona um item no final da lista
#nomes[1] = 'Kuririn' #altera o item na posição 1 da lista
nomes.insert(1, 'Kuririn') #adiciona um item na posição 1 da lista
nomes.remove('Goku') #remove o item 'Goku' da lista
nomes.sort() #ordena a lista em ordem alfabética
print(nomes) #printa a lista completa
```

# Coleção

```python
nomes = {'Goku', 'Vegeta', 'Trunks', 'Gohan', 'Trunks', 'Goku'} #coleção(sem ordem definida e sem itens repetidos)
#os itens não possuem índice, então se eu fizer print(nomes[1]) dará erro
nomes.add('Bulma') #adiciona um item na coleção
nomes.remove('Gohan') #remove o item 'Gohan' da coleção
print(nomes) #printa a coleção completa (sem itens repetidos)
```

# Conjuntos

```python
A = {1, 2, 4}
B = {3, 4, 6}

Uniao = A | B
Intersecao = A & B
Diferenca = A - B
print(Uniao)
print(Intersecao)
print(Diferenca)
```

# Dicionário

```python
dados = {'nome': 'Goku', 'idade': 43} #dicionário(coleção de items com chave e valor)
dados['sexo'] = 'M'
print(dados['sexo'])
print(dados['nome'])
print(dados.keys()) #printa as chaves do dicionário
```

```python
dados1 = {'nome': 'Goku', 'idade': 43, 'sexo': 'M'} #dicionário(coleção de items com chave e valor)
dados2 = {'nome': 'Vegeta', 'idade': 45, 'sexo': 'M'} #dicionário(coleção de items com chave e valor)
dados3 = {'nome': 'Bulma', 'idade': 40, 'sexo': 'F'} #dicionário(coleção de items com chave e valor)
dbz = [dados1, dados2, dados3] #lista de dicionários

print(dbz[0]['nome']) #printa o nome do primeiro dicionário da lista (como se fosse uma matriz)
print(dbz[1]['idade']) #printa a idade do segundo dicionário da lista
print(dbz[2]['sexo']) #printa o sexo do terceiro dicionário da lista
```

# Exercícios

```python
lista = [] #inicializa uma lista vazia
for i in range(5):
    nome = input()
    lista.append(nome) #adiciona o nome digitado na lista
    
time_procurado = input()
print(f'Os 3 primeiros colocados sao: {lista[:3]}') #printa os 3 primeiros colocados
print(f'Os 2 ultimos colocados sao: {lista[-2:]}') #printa os 2 ultimos colocados
print(f'Times em oredem alfabetica: {sorted(lista)}') #printa a lista em ordem alfabetica

print(f'O {time_procurado} esta na posicao {lista.index(time_procurado) + 1}.') #printa a posicao do time procurado
```

```python
quantidade_loja_01 = int(input()) #número inteiro que representa a quantidade de produtos da loja 01
lista_loja_01 = [] #inicializa uma lista vazia para os produtos da loja 01

for i in range(quantidade_loja_01): #loop que vai de 0 até a quantidade de produtos da loja 01
    produto = input() #lê o nome do produto
    lista_loja_01.append(produto) #adiciona o produto digitado na lista da loja 01
    

conjunto_loja_01 = set(lista_loja_01) #converte a lista da loja 01 em um conjunto para evitar duplicatas


#----------------------------------------------------------------------------------------------------------------
quantidade_loja_02 = int(input()) #número inteiro que representa a quantidade de produtos da loja 02
lista_loja_02 = [] # inicializa uma lista vazia para os produtos da loja 02

for i in range(quantidade_loja_02): #loop que vai de 0 até a quantidade de produtos da loja 02
    produto = input() #lê o nome do produto
    lista_loja_02.append(produto) #adiciona o produto digitado na lista da loja 02

conjunto_loja_02 = set(lista_loja_02) #converte a lista da loja 02 em um conjunto para evitar duplicatas


# União e interseção
uniao = conjunto_loja_01 | conjunto_loja_02
intersecao = conjunto_loja_01 & conjunto_loja_02

# Imprimindo em ordem alfabética como lista
print(f"Modelos disponíveis em pelo menos uma das lojas: {sorted(uniao, key=str.lower)}")
print(f"Modelos disponíveis em ambas as lojas: {sorted(intersecao, key=str.lower)}")
```
