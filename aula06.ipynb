# Pandas Groupby

```python
import pandas as pd
```

```python
data = {
    'Categoria': ['A', 'A', 'B', 'B', 'C', 'A'],
    'Valor': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)
print(df)
```

# Divisão(split)

```python
group = df.groupby('Categoria')
```

# Aplicação(apply)

```python
group['Valor'].sum()
```

# Usando agg para várias operações

```python
print(df.groupby('Categoria')['Valor'].agg(['sum', 'mean', 'max']))
```

```python
#importa o csv
dfPaises = pd.read_csv('paises.csv', sep=';')

#agrupando os países por região e mostrando as quantidades de cada região
group_region = dfPaises.groupby('Region')
print(group_region.count()['Country'])

#agrupando os países por região e a soma da população de cada região
print(group_region.sum()['Population'])

#agrupando países por região e descrevendo estatísticas importantes
print(group_region.describe())

```

# Criando métodos customizados

```python

#retorna 90% do valor de x
def tenpercent(x):
    return x*0.9

#mostrando a taxa de mortalidade dos países
deathrate1 = dfPaises['Deathrate']

#aplicando o método por meio da função apply()
#diminuindo a taxa de mortalidade dos países em 10%
deathrate2 = dfPaises['Deathrate'].apply(tenpercent)

#concatenando e mostrando as duas series
pd.concat([deathrate1, deathrate2], axis=1)
```

```python
import pandas as pd

# 'index_col=0' diz ao pandas para usar a primeira coluna do arquivo (W) como índice.
# Sem isso, A,B,C,D,E seriam apenas valores de uma coluna normal e não dariam para acessar com .loc["D"]
arquivo = input() #entrando com o nome do arquivo para poder selecionar qual dos dois eu quero gerenciar dados
dfDados = pd.read_csv(arquivo, index_col=0)

# filtrar os valores menores que 30
valores_maiores_30 = dfDados[dfDados['X'] > 30]['X']
#dfDados['X'] < 30   ->  if(x < 30), seta true na linha se for verdade, e false se for maior que 30
#dfDados[dfDados['X'] < 30]    -> mostra em dataframe as linhas que eram true(no caso, eram B e C), então vai mostrar linhas e colunas
#quando eu coloco ['X'], eu peço para mostrar só a coluna X das linhas onde eram true

media = valores_maiores_30.mean()
print(f"Media dos elementos maiores que 30 da coluna X:  {media}")

#----------------------------------------------------------------------------------------------------------

# pegar a linha D
linha_D_loc = dfDados.loc["D"]

# calcular a média
media_D_loc = linha_D_loc.mean()

print("Media da linha D com loc:", media_D_loc)

#----------------------------------------------------------------------------------------------------------

#pegar a linha D
linha_D_iloc = dfDados.iloc[3,0:]

#calcular a média
media_D_iloc = linha_D_iloc.mean()

print(f"Media da linha D com iloc: {media_D_iloc}")

#----------------------------------------------------------------------------------------------------------

print("Dataframe reduzido:")
slicing = dfDados.loc[['A', 'C', 'E'],['X', 'Y']] #filtra o dataframe com base em rótulos(nomes)

print(slicing)

#----------------------------------------------------------------------------------------------------------

print("Medias das colunas:")
medias_colunas = slicing.mean(axis=0)

print(medias_colunas)

#----------------------------------------------------------------------------------------------------------

print("Medias das linhas:")

medias_linhas = slicing.mean(axis=1)

print(medias_linhas)
```

```python
import pandas as pd
#-----------------------------------------------------------------------------------------------------------------

contador = 0
# importa o csv
dfPaises = pd.read_csv('paises.csv', sep=';')

# remove espaços em branco na coluna 'Region'
dfPaises['Region'] = dfPaises['Region'].str.strip()

# agrupa por região
group_region = dfPaises.groupby('Region')

oceania = group_region.get_group('OCEANIA') #dataframe 

print('Países da OCEANIA:')
for pais in oceania['Country']: #isso tudo é como se fosse essa query: SELECT Country FROM paises WHERE Region = 'OCEANIA';
    contador+= 1
    print(pais)
    
#-----------------------------------------------------------------------------------------------------------------

print() #pula uma linha

print(f"Quantidade de países da OCEANIA: {contador}")

#-----------------------------------------------------------------------------------------------------------------

print() #pula uma linha

linha_max = dfPaises.loc[dfPaises['Population'].idxmax()] #seleciona a linha com maior número

print(f"País com maior população: {linha_max['Country']}")
print(f"Região: {linha_max['Region']}") 

#-----------------------------------------------------------------------------------------------------------------
print()

print("Média de alfabetização por região:")

# calcula a média de alfabetização por região e arredonda para 2 casas decimais
media_literacy = dfPaises.groupby('Region')['Literacy (%)'].mean()

# imprime cada região com a média formatada
for regiao, valor in media_literacy.items():
    print(f"{regiao:<30}: {valor:.2f}%")
    
    
#-----------------------------------------------------------------------------------------------------------------

print()

print('Países sem costa marítima:')

# filtra países sem costa (dataframe)
paises_sem_costa = dfPaises[dfPaises['Coastline (coast/area ratio)'] == 0]

# imprime o nome dos países
for pais in paises_sem_costa['Country']:
    print(pais)
```
